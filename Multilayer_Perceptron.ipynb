```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
# shufffle the  data
df.sample(frac = 1)
```

```python
# one hot encoding using pandas library
# we have to  change the value of species into numerical value

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python


df['species'] = le.fit_transform(df['species'])
```

```python
df
```

```python
X = df.drop(columns = 'species' , axis = 1)
```

```python

y = df['species']
```

```python
X
```

```python
y
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X, y ,test_size = 0.2, random_state = 20)
```

```python
# now create our model ----> Multilayer Perceptron
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model  = Sequential()

model.add(Dense(6 , activation = 'sigmoid'))   #  hidden layer
model.add(Dense(3 , activation = 'softmax')) # output layer
model.compile(loss = 'sparse_categorical_crossentropy', metrics  = ["accuracy"])
model.fit(X_train , y_train, epochs = 25 , batch_size = 5)
```

```python
model.evaluate(X_test , y_test)
```
