```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow 
from tensorflow import keras
from tensorflow.keras.layers import Dense
import keras
from keras.models import Sequential
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples = 250 , noise = 0.05 , random_state = 20)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X[: ,0] , X[: ,1] , c =y , s =100)
plt.show()
```

```python
# model building
model = Sequential()
model.add(Dense(10 , activation = 'sigmoid' , input_dim  =2))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(10 , activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = "Adam" , metrics = ['accuracy'])
```

```python
# weights of layers
model.get_weights()[0]

old_weights = model.get_weights()[0]
old_weights
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y ,test_size = 0.2 , random_state = 20)
```

```python
model.fit(X_train , y_train , epochs = 1)
```

```python
new_weights =  model.get_weights()[0]
new_weights
```

```python
# now we can calculate gradient = gradient = (old_weight-new_weights)/0.01

gradient = (old_weights - new_weights) / 0.01
gradient
```

```python
# we can clearly see that our gradient is converging to zero and the change in old weight and new wieght after training in 
# negligible

# so there is vanishing gradient problem
```
